**Question:** Which players are most undervalued by xG alone when you add on-ball value to the picture?

**Data:** StatsBomb Open Data — OBV calculable from event data

**Why this question matters:**  
xG captures shot quality but ignores everything a player does before the shot. A midfielder who consistently moves the ball into dangerous areas, creates space through carrying, and finds passes between lines may generate little xG directly but enormous value through the actions that enable others to shoot. On-Ball Value (OBV) attempts to capture this by valuing every action based on how it changes the probability of scoring. Players who rank highly on OBV but lowly on xG are candidates for systematic undervaluation.

## Step 1: Understand On-Ball Value (OBV)

Before writing any code, understand what OBV measures.

StatsBomb's OBV framework assigns a value to every on-ball action based on how it changes the probability of the team scoring or conceding in the next possession. A pass that moves the ball into a dangerous zone has positive OBV. A misplaced pass that gives away possession has negative OBV.
