**Question:** Which forwards have the largest gap between non-penalty xG and actual goals — consistently across three or more seasons?

**Data:** FBref + Understat — multi-season

**Why this question matters:**  
Single-season xG overperformance is mostly noise. A forward who beats their xG by 3 goals in one season is probably lucky. A forward who beats it consistently across three or more seasons is probably a better finisher than the model assumes. Finding these players — and understanding why the model undervalues them — is one of the most practically useful questions in recruitment analytics.
