**Question:** How does passing network structure change when a team is winning versus losing?

**Data:** StatsBomb Open Data — La Liga (Barcelona seasons)

**Why this question matters:**  
Passing networks reveal how a team's structure and connectivity changes under different conditions. A team winning 2-0 may become more conservative, channeling passes through fewer players. A team losing may open up, creating more connections and higher risk. Understanding this helps analysts identify tactical flexibility — and fragility.
