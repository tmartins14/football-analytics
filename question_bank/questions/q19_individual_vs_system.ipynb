**Question:** How much of a player's possession value metrics are attributable to the individual versus the system — using player transfers as natural experiments?

**Data:** FBref player stats 2017-2024 + Transfermarkt transfer records

**Why this question matters:**  
The system vs individual attribution problem is the most important unsolved question in football recruitment analytics. When a player moves clubs and their metrics change, how much is the player and how much is the system? Transfers provide a natural experiment: the player's individual quality is held constant while the system changes. Exploiting this structure is one of the most sophisticated things an analyst can do with publicly available data.

**What you are building toward:**
- A transfer-based natural experiment framework
