**Question:** Does pressing intensity cause improved defensive outcomes, or does team quality confound the relationship?

**Data:** FBref — PPDA and defensive metrics + Transfermarkt squad values as quality proxy

**Why this question matters:**  
The correlation between pressing intensity and good defensive outcomes is well documented. But high-quality squads both press more effectively AND concede less for reasons unrelated to pressing. Without controlling for squad quality, the pressing-defense relationship is confounded. This question introduces the core causal inference problem that distinguishes serious analysts from those who mistake correlation for causation.
