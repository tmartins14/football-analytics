```python
import sys
from pathlib import Path

# Add project root so we can import from src/statsbomb/
project_root = Path().resolve().parents[1]
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.statsbomb.extract_shots import resolve_euro_2024_final, extract_shots
import pandas as pd
import json
from pathlib import Path
```
